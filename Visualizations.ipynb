```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Reading Data File
titanic_data_file_path = 'train.csv'
titanic_data = pd.read_csv(titanic_data_file_path)
```

```python
titanic_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Sex vs. Survived")

# Bar chart 
sns.boxplot(y=titanic_data['Sex'], x=titanic_data['Survived'])

# Add label for vertical axis
plt.ylabel('Survived')
```

```python
plt.figure(figsize=(10,6))

# Add title
plt.title("Age vs. Survived")

# Bar chart 
sns.barplot(x=titanic_data['Survived'], y=titanic_data['Age'])

# Add label for vertical axis
plt.ylabel('Age')
```
